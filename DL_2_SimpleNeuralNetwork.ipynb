```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
model = Sequential ()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
#model.add(Dense(32, activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Where 1500/1500 comes from while training.<br>
Total samples = 60,000.<br>
validation_split = 0.2 → training samples = 48,000 <br>
batch_size = 32 <br>
steps_per_epoch = 48,000 / 32 = 1500

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

```python
history = model.fit(X_train, y_train, epochs=25, validation_split=0.2)
```

```python
model.predict(X_test)
```

We store it in a variable in y probability

```python
y_prob = model.predict(X_test)
```

First test image is 7, then second as 2, third as 1 and so on.

```python
y_prob.argmax(axis=1)
```

```python
y_pred = y_prob.argmax(axis=1)
```

Accuracy below

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
